```python
X = df.iloc[:, 1:]
Y = df.iloc[:, 0]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(12,10))
corr = X_train.corr()
sns.heatmap(corr, annot=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
from sklearn.metrics import  r2_score
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
lr_model = LinearRegression()
lr_model.fit(X_train_scaled, Y_train)
y_pred1 = lr_model.predict(X_test_scaled)
print(r2_score(Y_test,y_pred1))
```

```python
model_ridge = Ridge(alpha=1)
model_ridge.fit(X_train_scaled, Y_train)
y_pred2 = model_ridge.predict(X_test_scaled)
print(r2_score(Y_test,y_pred2))
```

```python
model_lasso = Lasso()
model_lasso.fit(X_train_scaled, Y_train)
y_pred3 = model_lasso.predict(X_test_scaled)
print(r2_score(Y_test,y_pred3))
```

```python
model_elastic = ElasticNet()
model_elastic.fit(X_train_scaled, Y_train)
y_pred4 = model_elastic.predict(X_test_scaled)
print(r2_score(Y_test,y_pred4))
```

```python
model5 = DecisionTreeRegressor()
model5.fit(X_train_scaled, Y_train)
y_pred5 = model5.predict(X_test_scaled)
print(r2_score(Y_test,y_pred5))
```

```python
model6 = RandomForestRegressor()
model6.fit(X_train_scaled, Y_train)
y_pred6 = model6.predict(X_test_scaled)
print(r2_score(Y_test,y_pred6))
```

```python
from sklearn.linear_model import LassoCV
lassocv = LassoCV(cv = 5)
lassocv.fit(X_train, Y_train)
```

```python
y_pred_lassocv = lassocv.predict(X_test_scaled)
```
